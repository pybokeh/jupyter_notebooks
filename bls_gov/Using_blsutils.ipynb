# Using `blsutils` package

This notebook demonstrates how to use my custom `blsutils` package (not on pypi) to plot various BLS series.  You need to have your own bls developer API key and also have the series ID you want to plot.

```python
from blsutils import plot_bls_series_id
from pathlib import Path
import configparser
```

```python
config = configparser.ConfigParser()
config.read(Path.home() / '.config' / 'config.ini')
bls_key = config['bls']['secretkey']
```

#### Plotting last 19 years by default (maximum allowed by the BLS API)

```python
plot_bls_series_id(
    series_id='CES6000000001',
    series_descr='Number of All employees, thousands, professional and business services, seasonally adjusted',
    bls_key=bls_key
)
```

#### Instead of plotting last 19 years' worth of data, you can specify the last n years worth of data you want to plot.  But in doing so, the shaded regions indicating recessions and COVID-19 pandemic will not be created.

```python
plot_bls_series_id(
    series_id='CES6000000001',
    series_descr='Number of All employees, thousands, professional and business services, seasonally adjusted',
    bls_key=bls_key,
    past_n_years=5
)
```

It does appear that hiring has slowed down during summer of 2023 and has slightly picked up recently.
