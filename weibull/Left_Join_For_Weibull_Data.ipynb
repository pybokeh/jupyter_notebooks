```python
import pandas as pd

url = input("Paste the URL\path to your data here: ")
    
left_table = pd.read_excel(url, input("Enter left worksheet name: "), index_col=0)
print("*********  Completed parsing left worksheet  *********\n")
right_table = pd.read_excel(url, input("Enter right worksheet name: "), index_col=0)
print("*********  Completed parsing right worksheet  *********")

left_joined = pd.merge(left_table, right_table, how='left', left_index=True, right_index=True)
left_joined = left_joined[['DTF', 'MTF', 'DAYS_TO_FAIL_MINZERO', 'MILES_TO_FAIL']]

# Below are functions used to create Status column, update the DTF and MTF columns
def createStatus(series):
    if pd.isnull(series['DAYS_TO_FAIL_MINZERO']):
        return 'SUSPENDED'
    else:
        return 'FAILED'
    
def updateDTF(series):
    if pd.isnull(series['DAYS_TO_FAIL_MINZERO']) and series['DTF'] > 0:
        return series['DTF']
    else:
        return series['DAYS_TO_FAIL_MINZERO']
    
def zero2oneDTF(series):
    if series['DTF'] == 0:
        return 1
    else:
        return series['DTF']

def updateMTF(series):
    if pd.isnull(series['MILES_TO_FAIL']):
        return series['MTF']
    else:
        return series['MILES_TO_FAIL']

left_joined['STATUS'] = left_joined.apply(createStatus, axis=1)
left_joined['DTF'] = left_joined.apply(updateDTF, axis=1)
left_joined['MTF'] = left_joined.apply(updateMTF, axis=1)
left_joined['DTF'] = left_joined.apply(zero2oneDTF, axis=1)

left_joined = left_joined.drop(['DAYS_TO_FAIL_MINZERO','MILES_TO_FAIL'], axis=1)
```

```python
left_joined.sort(['STATUS','DTF'], axis=0, inplace=True)
```
