```python
import pandas as pd
import matplotlib.pyplot as plt

xls = pd.ExcelFile('path_to_file')
df = xls.parse('Claims', index_col=['CLM_SK'], na_values=['NA'])
```

```python
df = df[['DAYS_TO_FAIL_MINZERO','MILES_TO_FAIL']]
dtf = df['DAYS_TO_FAIL_MINZERO'].values
mtf = df['MILES_TO_FAIL'].values
```

```python
stats = df['DAYS_TO_FAIL_MINZERO'].describe()
count = int(stats[0])
mean = stats[1]
std = stats[2]
min = int(stats[3])
pct25 = stats[4]
pct50 = stats[5]
pct75 = stats[6]
max = int(stats[7])

stats_label = "count: " + str(count) + "\n" + "min: " + str(min) + "\n" + "max: " + str(max) + "\n" + "median: " + str(pct50)
```

```python
# http://matplotlib.org/users/text_intro.html
fig = plt.figure()
ax = fig.add_subplot(111)
plt.hist(dtf, bins=15)
plt.grid(True)
plt.xlabel("DTF")
axis = plt.gca()
ax.text(100, 100, stats_label, style='italic',
        bbox={'facecolor':'gray', 'alpha':0.7, 'pad':10})
plt.show()
```

```python
stats = df['MILES_TO_FAIL'].describe()
count = int(stats[0])
mean = stats[1]
std = stats[2]
min = int(stats[3])
pct25 = stats[4]
pct50 = stats[5]
pct75 = stats[6]
max = int(stats[7])

stats_label = "count: " + str(count) + "\n" + "min: " + str(min) + "\n" + "max: " + str(max) + "\n" + "median: " + str(pct50)
```

```python
#http://matplotlib.org/users/text_intro.html
fig = plt.figure()
ax = fig.add_subplot(111)
plt.hist(mtf, bins=20)
axis = plt.gca()
plt.grid(True)
plt.xlabel("MTF")
ax.text(80000, 120, stats_label, style='italic',
        bbox={'facecolor':'gray', 'alpha':0.7, 'pad':10})
plt.show()
```
