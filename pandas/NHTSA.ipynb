#### Now that we've examined the data, let's create a sqlite table from the dataframe.&nbsp;&nbsp;For those with limited RAM, this method should work OK with large files.

```python
import sqlite3
import codecs

conn = sqlite3.connect(r'D:\webapps\denso\server\db\nhtsa')
c = conn.cursor()

with codecs.open(r'D:\temp\NHTSA\FLAT_RCL.txt','r', encoding="ISO-8859-1") as f:
    for line in f:
        delimited_data = line.split('\t')
        c.execute('INSERT INTO recalls VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)', delimited_data)
        conn.commit()
c.close()
conn.close()
```
